```python
#preparing Model
import statsmodels.formula.api as smf
model = smf.ols('Price~Age+KM+HP+Doors+Gears+Weight',data=toyoto).fit()
```

```python
model.params
```

```python
print(model.tvalues,'\n',model.pvalues)
(model.rsquared,model.rsquared_adj)
```

```python
#Simple Linear Regression Models
#model doors
ml_d = smf.ols('Price~Doors',data=toyoto).fit()
```

```python
print(ml_d.tvalues,'\n',ml_d.pvalues)
(ml_d.rsquared,ml_d.rsquared_adj)
```

```python
#Calculation VIF
rsq_age = smf.ols('Age~KM+HP+Doors+Gears+Weight',data=toyoto).fit().rsquared
vif_age = 1/(1-rsq_age)

rsq_km = smf.ols('KM~Age+HP+Doors+Gears+Weight',data=toyoto).fit().rsquared
vif_km = 1/(1-rsq_km)

rsq_hp = smf.ols('HP~KM+Age+Doors+Gears+Weight',data=toyoto).fit().rsquared
vif_hp = 1/(1-rsq_hp)

rsq_doors = smf.ols('Doors~KM+HP+Age+Gears+Weight',data=toyoto).fit().rsquared
vif_doors = 1/(1-rsq_doors)

rsq_gears = smf.ols('Gears~KM+HP+Doors+Age+Weight',data=toyoto).fit().rsquared
vif_gears = 1/(1-rsq_gears)

rsq_weight = smf.ols('Weight~KM+HP+Doors+Gears+Age',data=toyoto).fit().rsquared
vif_weight = 1/(1-rsq_weight)

d1 = {'Variables':['Age','KM','HP','Doors','Gears','Weight'],
      'VIF':[vif_age,vif_km,vif_hp,vif_doors,vif_gears,vif_weight]}
vif_frame = pd.DataFrame(d1)
vif_frame
```

```python
#Model Deletion Diagnostics
#Detecting Influencers/Outliers
#Cook’s Distance
from statsmodels.graphics.regressionplots import influence_plot

model_influence = model.get_influence()
(c,_)=model_influence.cooks_distance
```

```python
fig = plt.subplots(figsize=(20,7))
plt.stem(np.arange(len(toyoto)), np.round(c,3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
(np.argmax(c),np.max(c))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
uni = pd.read_csv('Copy of Universities.csv')
uni.head()
```

```python
uni_df= uni.iloc[:,1:]
uni_df.head()
```

```python
scaler = StandardScaler()
scaler.fit_transform(uni_df)
```

```python
pca_2 = PCA(n_components=2)
pca_components = pd.DataFrame(pca_2.fit_transform(uni_df),columns=['a','b'])
pca_components.head()
```

```python
plt.scatter(pca_components.a,pca_components.b)
```

```python
pca_3 = PCA(n_components=3)
pca_components3 = pd.DataFrame(pca_3.fit_transform(uni_df),columns=['a','b','c'])
```

```python
pca_components3
```

```python
pca = PCA(n_components = 6)
pca_values = pca.fit_transform(uni_df)
```

```python
pca_components6 = pd.DataFrame(pca.fit_transform(uni_df),columns=['a','b','c','d','e','f'])
pca_components6
```

```python
pca_values
```

```python
variance = pca.explained_variance_ratio_
variance
```

```python
pca.components_
```

```python
plt.plot(variance, color='red')
plt.title('Scree plot')
plt.show()
```

```python
x = pca_values[:,0]
y = pca_values[:,1]
plt.scatter(x,y)
```

```python
finalDf = pd.DataFrame(pca_values[::,0:2], columns=['pc1','pc2'])
```

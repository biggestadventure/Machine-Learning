```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
x = [-3, 5, 7]
y = [10, 2, 5]
```

```python
plt.figure(figsize=(15,5))
plt.plot(x,y)
plt.xlabel('Sales',size=15,color='red')
plt.ylabel('Product',size=15,color='red')
plt.xlim(0,10)
plt.ylim(-3,8)
plt.title('Line Plot',size=15)
plt.suptitle('Sales Comparison', size=20, y=1)
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(2,2)
ax[0,0].set_title('Plot 1')
ax[0,0].plot(x,y)
ax[0,1].hist(x)
ax[1,0].pie(y)
ax[1,1].boxplot(x)

plt.show()
```

```python
fig, ax = plt.subplots(2,2,figsize=(10,10))
ax[0,0].plot(x,y,linestyle='dashed',color='green')
ax[0,0].set_title('Line',size=15)
ax[0,1].pie(y)
ax[0,1].set_title('Piechart',size=15)
ax[1,0].scatter(x,y,color='red')
ax[1,0].set_title('Scatter Plot',color='orange',size=15)
ax[1,1].hist(y,color='blue')
ax[1,1].set_title('Histogram',size=15)
plt.show()
```

## seaborn

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## Univariate --> "One variable" data visualization

#### Strip plot

```python
sns.stripplot(x='sex',y='tip', data=tips, jitter=True)
plt.title('Strip plot')
plt.xlabel('Gender')
plt.show()
```

```python
#Grouping with stripplot()
sns.stripplot(x='sex',y='tip',data=tips,hue='day')
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.stripplot(x='day',y='tip',data=tips,jitter=False,size=4)
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.stripplot(x='day',y='tip',data=tips,hue='sex',jitter=True,size=4)#without jitter = True, showing the same output
plt.ylabel('tip ($)')
plt.show()
```

### Swarm plot

```python
sns.swarmplot(x='day',y='tip',data=tips,hue='sex')
plt.ylabel('tip ($)')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.swarmplot(x='day',y='tip',data=tips,facecolor='black')
sns.violinplot(x='day',y='tip',data=tips,color='black')
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.violinplot(x='sex',y='tip',data=tips,color='red')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.swarmplot(x='sex',y='tip',data=tips,facecolor='black')
sns.violinplot(x='sex',y='tip',data=tips,color='black')
plt.ylabel('tip ($)')
plt.show()
```

```python
#changing orientation
sns.swarmplot(x='tip',y='day',data=tips,hue='sex',orient='h')
plt.xlabel('tip ($)')
plt.show()
```

```python
sns.barplot(x='sex',y='tip',data=tips)
plt.show()
```

```python
sns.histplot(x='day',y='tip',data=tips,hue='sex')
plt.show()
```

### Box and Violin plot

```python
#plt.subplot(1,2,1)
plt.figure(figsize=(10,10))
sns.violinplot(x='day',y='tip',data=tips)
sns.boxplot(x='day',y='tip',data=tips)

plt.ylabel('tip ($)')
plt.show()
```

```python
#plt.subplot(1,2,2)
sns.violinplot(x='day',y='tip',data=tips)
plt.ylabel('tip ($)')
plt.show()
```

```python
fig, ax=plt.subplots(2,2)
plt.subplot(2,2,1)
sns.violinplot(x='day',y='tip',data=tips)
plt.ylabel('tip ($)')


plt.subplot(2,2,2)
sns.violinplot(x='day',y='tip',data=tips)
sns.boxplot(x='day',y='tip',data=tips)
plt.ylabel('tip ($)')

plt.subplot(2,2,3)
sns.stripplot(x='sex',y='tip',data=tips,hue='day')
plt.ylabel('tip ($)')

plt.subplot(2,2,4)
sns.swarmplot(x='day',y='tip',data=tips,hue='sex')
plt.ylabel('tip ($)')
plt.show()
```

### Bivariate --> "two variables" data visualization

```python
#join plot
sns.jointplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
tips.head()
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips)
plt.show()
```

```python
#density plot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')#kernal density estimate
plt.show()
#by more distribution,more circle rounded together in centere
```

```python
#kind{ “scatter” | “kde” | “hist” | “hex” | “reg” | “resid” }
#reg=regplot() : This method is used to plot data and a linear regression model fit
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
#scatter
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
plt.show()
```

```python
#hist
sns.jointplot(x='total_bill', y= 'tip', data=tips, kind='hist')
plt.show()
```

```python
#hex
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
plt.show()
```

```python
#resid
sns.jointplot(x='total_bill',y='tip',data=tips,kind='resid')
plt.show()
```

## Titanic data Visualization

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('titanic.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

### Proportion of Survived

```python
data.groupby('Survived')['PassengerId'].count()
```

```python
data['Survived'].value_counts()
```

### Visual exploration

```python
plt.figure(figsize=(11,5))
sns.boxplot(x='Survived',y='Age',data=data); #Why ';' we using instead of plt.show() or by removing the descriptn
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x='Sex', y='Survived',data = data)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='Sex', y='Age',data=data, hue='Survived');
```

```python
sns.barplot(x='Pclass', y='Survived', data=data);
```

```python
sns.barplot(x='Pclass',y='Survived',hue='Sex',data=data);
```

```python
sns.barplot(x='SibSp',y='Survived',data=data);
```

```python
sns.barplot(x='Parch',y='Survived', data=data);
```

```python
survived = data.loc[data['Survived']==1,'Age'].dropna()
sns.distplot(survived)
plt.title('Survived');
```

```python
data.head()
```

```python
survived = data.loc[data['Survived']==1, 'Pclass'].dropna()
sns.distplot(survived)
plt.title('SURVIVED');
#in dist plot should only 1 variable
```

```python
not_survived = data.loc[data['Survived']==0,'Age'].dropna()
sns.distplot(not_survived)
plt.title('NOT SURVIVED');
```

```python
sns.pairplot(data);
```

```python
sns.distplot(a=data[data['Embarked']=='C']['Survived'],bins=3,kde=False)#kde=False means no need the density line
plt.title('Charboug')
plt.xticks([0,1])
plt.show()

sns.distplot(a=data[data['Embarked']=='Q']['Survived'],bins=3,kde=False)#kde=False means no need the density line
plt.title('QueenTown')
plt.xticks([0,1])
plt.show()

sns.distplot(a=data[data['Embarked']=='C']['Survived'],bins=3,kde=False)#kde=False means no need the density line
plt.title('Southampton')
plt.xticks([0,1])
plt.show()
#Most of the Passengers embarked from Southampton
```

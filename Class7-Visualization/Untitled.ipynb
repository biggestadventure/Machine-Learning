```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
x = [-3,5, 7]
y = [10, 2, 5]
```

```python
x
```

```python
plt.figure(figsize=(10,5))
plt.plot(x,y)
plt.xlabel('Sales', size=15, color='red')
plt.ylabel('Product', size=15, color='red')
plt.xlim(-3,10)
plt.ylim(-3,10)
plt.title('Line Plot', size=15)
plt.suptitle('Sales Comparison', size=20, y=1)
plt.legend()
plt.show()
```

```python
fig, ax=plt.subplots(2,2)
ax[0,0].set_title('Plot 1')
ax[0,1].set_title('Plot 2')
ax[1,0].set_title('Plot 3')
ax[1,1].set_title('Plot 4')
ax[0,0].plot(x,y)
ax[0,1].hist(x)
```
